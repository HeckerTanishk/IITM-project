```python
import pandas as pd
df = pd.read_csv('Book1.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import statistics as st
mean = st.mean(df['marks'])
print(f"The mean of the marks is: {mean}")
mode = st.mode(df['marks'])
print(f"The mode of the marks is: {mode}")
median = st.median(df['marks'])
print(f"The median of the marks is: {median}")
```

```python
sd = st.stdev(df['marks'])
print(f"The standard deviation of the marks is: {sd}")
variance = st.variance(df['marks'])
print(f"The variance of the marks is: {variance}")
```

```python
plt.hist(df['marks'], bins=5, edgecolor='black')
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='g', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='b', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram of Marks')
plt.legend()
plt.show()
```

```python
#total marks
df['total'] = df['maths'] + df['science'] + df['Engliash'] + df['SSt']
df['Average'] = df['total']/4

def get_performance(average):
  if average <= 35:
    return 'fail'
  elif 35 < average <= 50:
    return 'need improvement'
  elif 50 < average <= 60 :
    return 'Average'
  elif 60 < average <= 70:
    return 'good'
  elif 70 < average <= 80:
    return 'very good'
  else:
    return 'excellent'

df['performance'] = df['Average'].apply(get_performance)
display(df.head())
```

```python
plt.bar(df['name'], df['total'], edgecolor='black')
plt.xlabel('Student Name')
plt.ylabel('Total Marks')
plt.title('Total Marks per Student')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
subject_averages = df[['maths', 'science', 'Engliash', 'SSt']].mean()
average_marks_df = pd.DataFrame({
    'Subject': subject_averages.index,
    'Average Marks': subject_averages.values
})
plt.plot(average_marks_df['Subject'], average_marks_df['Average Marks'], marker='o', linestyle='-', color='b')
plt.xlabel('Subject')
plt.ylabel('Average Marks')
plt.title('Average Marks per Subject')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
